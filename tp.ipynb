```python
import numpy as np

def count_elements(arr):
    unique_elements, counts = np.unique(arr, return_counts=True)
    return dict(zip(unique_elements, counts))

array = np.array([1, 2, 3, 2, 4, 5, 6, 4, 7])
element_counts = count_elements(array)
print(element_counts)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Courbe de la fonction sinus')
plt.xlabel('X')
plt.ylabel('Sin(X)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = x
y2 = x ** 2

plt.plot(x, y1, label='Linéaire')
plt.plot(x, y2, label='Quadratique')

plt.title('Courbes linéaire et quadratique')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
